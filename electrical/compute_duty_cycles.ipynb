```python
import pandas as pd
df = []
def print_duty_cycles():
        i = 0
        while i <= 100:
            duty_small  = (i-0.0314)/0.0169
            duty_large = (i-0.044)/0.0236
            if duty_small > 100 or duty_large > 100:
                  break
            # print(f"Num:{i}, Duty Cycle Large: {duty_large}, Duty Cycle Small: {duty_small}")
            df.append([i,round(duty_large, 2),round(duty_small, 2)])
            i += 0.1
print_duty_cycles()
df = pd.DataFrame(df)
df.columns = ['Linear Velocity', 'Large Wheel Duty Cycle', 'Small Wheel Duty Cycle']
```
