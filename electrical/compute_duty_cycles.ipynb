```python
import pandas as pd
df = []
def print_duty_cycles():
        i = 0
        while i <= 100:
            duty_small  = (i-0.0942)/0.0506
            duty_large = (i-0.132)/0.0709
            if duty_small > 100 or duty_large > 100:
                  break
            # print(f"Num:{i}, Duty Cycle Large: {duty_small}, Duty Cycle Small: {duty_large}")
            df.append([i,round(duty_large, 2),round(duty_small, 2)])
            i += 0.5
print_duty_cycles()
df = pd.DataFrame(df)
df.columns = ['Constant', 'Duty Cycle Small', 'Duty Cycle Large']
```
